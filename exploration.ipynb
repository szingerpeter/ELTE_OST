```python
def print_structure(obj, path = ""):
    if type(obj) == h5py._hl.dataset.Dataset:
        print(f"Dataset after path: {path}\nValues:\n{obj[()]}\ntype:{type(obj[()])}\nshape:{obj.shape}\n*****")
    elif type(obj) == h5py._hl.group.Group or type(obj) == h5py._hl.files.File:
        for key in obj.keys():
            sub_obj = obj[key]
            print_structure(sub_obj, f"{path}/{key}")
    else:
        print("Unexpected type:", type(obj))
```

```python
#adapt
sample = h5py.File(main_dir + "adapt/X1.h5")
print_structure(sample)
```

```python
def get_values_from_objects(dir_path, val_path, transformation = lambda x: x):
    tmp = {}
    for fn in os.listdir(dir_path):
        obj = h5py.File(dir_path + fn)
        tmp[fn] = [transformation(x) for x in obj[val_path][()]]
    return tmp
```

```python
dates = get_values_from_objects(main_dir + "adapt/", "X/value/t/value", datetime.fromtimestamp)
date_values = [y for x in dates.values() for y in x]
plt.plot(sorted(date_values), range(len(date_values)))
plt.xticks(rotation = 45)
```

date range: 2010-07-01 - 2010-09-15 (~2.5 months)
