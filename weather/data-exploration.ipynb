```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
```

```python
df = pd.read_csv('data/data.txt', sep = '\s+')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
df = df.sort_values('date', ascending=True)
plt.plot(df['date'], df['max'])
plt.xticks(rotation='vertical')
```

```python
fig,ax = plt.subplots()

df = df.loc[(df['date'] > datetime.datetime(2019, 1, 1)) & (df['date'] < datetime.datetime(2020, 1, 1))]

ax.plot(df['date'], df['max'])
ax.plot(df['date'], df['min'])
```
