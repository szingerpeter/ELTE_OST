```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
```

```python
df = pd.read_csv('data/data.txt', sep = '\s+')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
df = df.sort_values('date', ascending=True)
plt.plot(df['date'], df['max'])
plt.xticks(rotation='vertical')
```

```python
plt.style.use("seaborn-dark-palette") 

fig,ax = plt.subplots()

df = df.loc[(df['date'] > datetime.datetime(2019, 1, 1)) & (df['date'] < datetime.datetime(2020, 1, 1))]

ax.plot(df['date'], df['max'], label='max')
ax.plot(df['date'], df['min'], label='min')

plt.title("Min/Max temperature along the year 2019 in Madrid")

plt.show()
```

```python
plt.style.use("seaborn-dark-palette")

# Histogram
heights,bins = np.histogram(df['max'],bins=50)

plt.plot(bins[:-1],heights)

plt.title("Maximum temperature distribution")

plt.show()

# Histogram
heights,bins = np.histogram(df['min'],bins=50)

plt.plot(bins[:-1],heights)

plt.title("Minimum temperature distribution")

plt.show()
```
